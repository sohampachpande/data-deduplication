```python
import pandas as pd
import numpy as np
```

```python
path = '../../data/error_analysis_RF_state_abb.csv'
df = pd.read_csv(path, usecols=['w1', 'w2', 'ground_truth', 'prediction'])
```

```python
df['w1'] = df['w1'].apply(lambda x: str(x))
df['w2'] = df['w2'].apply(lambda x: str(x))

df['len w1'] = df['w1'].apply(lambda x: len(x))
df['len w2'] = df['w2'].apply(lambda x: len(x))
```

```python
df['isCorrect'] = df['ground_truth']==df['prediction']
```

```python
def state_abv(x):
    if x.lower().strip() in statesDic:
        return statesDic[x.lower().strip()] 
    else:
        return x
df['w1'] = df['w1'].apply(lambda x: state_abv(x))
df['w2'] = df['w2'].apply(lambda x: state_abv(x))
```
